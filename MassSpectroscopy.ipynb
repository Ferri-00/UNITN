### Miscela

```python
I1 = miscela[34][1] 
I12 = miscela[359][1]
I13 = miscela[362][1]
I16 = miscela[372][1]
I26 = miscela[841][1]
I27 = miscela[868][1]
I28 = miscela[899][1]
```

```python
P1, P2, P3 = symbols('p_1,p_2,p_3')
```

```python
eq = [Eq(S1_1*P1 + S1_2*P2 + S1_3*P3, I1), 
      Eq(S12_1*P1 + S12_2*P2 + S12_3*P3,I12),
      Eq(S13_1*P1 + S13_2*P2 + S13_3*P3,I13),
      Eq(S16_1*P1 + S16_2*P2 + S16_3*P3,I16),
      Eq(S26_1*P1 + S26_2*P2 + S26_3*P3,I26),
      Eq(S27_1*P1 + S27_2*P2 + S27_3*P3,I27),
      Eq(S28_1*P1 + S28_2*P2 + S28_3*P3,I28)]
```

```python
result1 = solve([eq[3],eq[4],eq[5]], P1, P2, P3)
result2 = solve([eq[3],eq[4],eq[6]], P1, P2, P3)
result3 = solve([eq[3],eq[5],eq[6]], P1, P2, P3)
```
